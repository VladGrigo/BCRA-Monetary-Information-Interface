# Pasivos no Remunerados

```python
hist= pd.read_excel(file_url,header=[2,3,4,5], sheet_name='INSTRUMENTOS DEL BCRA')
```

```python
hist
```

```python
hist.columns = range(hist.shape[1])
hist
```

```python
hist[0] = pd.to_datetime(hist[0], errors='coerce')
hist = hist.dropna(subset=[0])
hist
```

```python
last_data_pas= hist[hist[0].dt.date == pd.to_datetime(date_of_last_data)]
```

```python
pasivo_rem= last_data_pas[[0,1,4,5,8]].copy()
pasivo_rem.tail()
```

```python
pasivo_rem.columns=["Fecha","Pases Pasivos en Pesos", "LELIQ", "LEBAC y NOBAC en Pesos y LEGAR","NOCOM"]
```

```python
pasivo_rem.sum()
```

```python
pasivo_rem= pasivo_rem[["Pases Pasivos en Pesos","LELIQ", "LEBAC y NOBAC en Pesos y LEGAR", "NOCOM"]].astype(float)
```

```python
pasivo_rem["Pasivos no Remunerados Total"]= pasivo_rem.sum(axis=1)
```

```python
rem_last_amount=pasivo_rem.iloc[-1]["Pasivos no Remunerados Total"]
```

# Creating Excel File 

```python
info_excel= pd.read_excel("BCRA Daily Monetary Information.xlsx")
```

```python
info_excel
```

```python
new_data={'Fecha ':date_of_last_data, 
          'Pas. Rem.':rem_last_amount, 
          'Base Mon.':monetary_base_last_amount, 
          '1+2':rem_last_amount + monetary_base_last_amount,
          "PR/BM":(rem_last_amount/monetary_base_last_amount)*100,
          "Reservas":reserves_last_amount,
          'TC ':exchange_rate_last_amount, 
          "TC+IMP":exchange_rate_last_amount*1.65, 
          "TC EQ":(rem_last_amount + monetary_base_last_amount)/reserves_last_amount
         }
```

```python
new_data
```

```python
if(date_of_last_data in info_excel["Fecha "].values):
    print("Oops! That data is already available in the file!")
else:
    info= info_excel.append(new_data, ignore_index=True)
    print("Saving the new data to existing file")
    info.to_excel('BCRA Daily Monetary Information.xlsx', index = False)
    print("Saved data to file!!")
```
